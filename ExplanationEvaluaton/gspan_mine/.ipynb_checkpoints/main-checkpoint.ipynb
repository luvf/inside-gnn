```python
def get_atom_map(dataset, label):
    if dataset == 'aids':
        atom_map = {0: 'C', 1: 'O', 2: 'N', 3: 'Cl', 4: 'F', 5: 'S', 6: 'Se', 7: 'P', 8: 'Na', 9: 'I', 10: 'Co',
                    11: 'Br', 12: 'Li', 13: 'Si', 14: 'Mg', 15: 'Cu', 16: 'As', 17: 'B', 18: 'Pt', 19: 'Ru', 20: 'K',
                    21: 'Pd', 22: 'Au', 23: 'Te', 24: 'W', 25: 'Rh', 26: 'Zn', 27: 'Bi', 28: 'Pb', 29: 'Ge', 30: 'Sb',
                    31: 'Sn', 32: 'Ga', 33: 'Hg', 34: 'Ho', 35: 'Tl', 36: 'Ni', 37: 'Tb'}
    if dataset == 'BBBP':
        atom_map = {0: 'C', 1: 'N', 2: 'O', 3: 'S', 4: 'P', 5: 'BR', 6: 'B', 7: 'F', 8: 'CL', 9: 'I', 10: 'H', 11: 'NA',
                    12: 'CA'}
    if dataset == 'mutag':
        atom_map = {0: 'C', 1: 'O', 2: 'Cl', 3: 'H', 4: 'N', 5: 'F', 6: 'Br', 7: 'S', 8: 'P', 9: 'I', 10: 'Na', 11: 'K',
                    12: 'Li', 13: 'Ca'}
    atom_map = {str(key): value for key, value in atom_map.items()}
    return atom_map[label]
```
